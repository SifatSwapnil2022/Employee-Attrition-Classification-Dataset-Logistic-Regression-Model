```python
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/lab exam/employee.csv')
# Counting occurrences of each category in the Education Level column
Associate_Degree = (df['Education Level'] == 'Associate Degree').sum()
Undergraduate = (df['Education Level'] == 'Bachelor’s Degree').sum()
High_School = (df['Education Level'] == 'High School').sum()
Postgraduate = (df['Education Level'] == 'Master’s Degree').sum()
PhD = (df['Education Level'] == 'PhD').sum()

# Proportions for each category
proportions = [Associate_Degree, Undergraduate, High_School, Postgraduate, PhD]

# Labels and colors with a gradient effect
labels = ['Associate Degree', 'Undergraduate', 'High School', 'Postgraduate', 'PhD']
colors = plt.cm.tab10(np.linspace(0, 1, len(labels)))  # Color map for a gradient effect

# Create a figure
plt.figure(figsize=(20, 12), dpi=100)

# Create a pie chart
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.pie(
    proportions,
    labels=labels,
    explode=(0.05, 0, 0, 0, 0),  # Slight explosion for the first slice
    startangle=90,
    autopct='%1.1f%%',
    shadow=True,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}  # Thicker edges for slices
)

plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.title("Education Level Proportion", fontsize=30, pad=20, fontweight='bold')

# Create a bar chart
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
bars = plt.bar(labels, proportions, color=colors, edgecolor='black', linewidth=1.5)
plt.ylabel('Number of Individuals', fontsize=20)
plt.title('Education Level Count', fontsize=30, pad=20, fontweight='bold')

# Add data labels on the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', fontsize=15, fontweight='bold')

# Style the x-axis
plt.xticks(rotation=45, fontsize=15)
plt.yticks(fontsize=15)

# Tight layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
#Grouped Bar Chart

import matplotlib.pyplot as plt
import numpy as np

# Sample data
education_levels = ['Master’s Degree', 'PhD', 'High School', 'Associate Degree','Bachelor’s Degree']
Job_Role = ['Education', 'Media', 'Healthcare','Technology','Finance']

# Count occurrences for each combination of education and income levels
education_income_counts = df.groupby(['Education Level', 'Job Role']).size().unstack(fill_value=0)

# Prepare data for plotting
education_counts = education_income_counts.loc[education_levels].T  # Transpose for easier plotting

# Plotting
fig, ax = plt.subplots(figsize=(14, 8), dpi=100)

# Bar width
bar_width = 0.2
index = np.arange(len(Job_Role))

# Plot bars for each education level
for i, education in enumerate(education_levels):
    bars = ax.bar(index + i * bar_width, education_counts[education], bar_width, label=education)

# Add labels, title, and legend
ax.set_xlabel('Job_Role Level', size=14)
ax.set_ylabel('Counts', size=14)
ax.set_title('Education Level Distribution by Job_Role Level', size=16, pad=20)
ax.set_xticks(index + bar_width * (len(education_levels) - 1) / 2)
ax.set_xticklabels(Job_Role)
ax.legend(title='Education Level')

# Show plot
plt.tight_layout()
plt.show()
```
